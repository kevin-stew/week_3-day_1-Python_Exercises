## Lambda Functions <br>
<p>Lambda functions... or "Anonymous Functions" are referring to inline functions with no name. The keyword lambda denotes the no name function, and executes within a single line. Without saving it to a variable; however, it is not able to be used, unless passed in either as a paramater or within list comprehension.<br>Written as "(keyword lambda) (one or more inputs) (colon) (function to be executed)"</p>

#### Syntax

```python
def addTwo(x):
    return x + 2

print(addTwo(6))

# Lambda Function Syntax
# Example of writing and calling a lambda function
print((lambda x : x + 2)(6))

```

#### Saving to a Variable

```python
f_test = lambda x : x + 2

f_test(8)
```

#### Multiple Inputs

```python
# multiple inputs no stored variable
print((lambda x,y,z: x * y * z)(2,4,10))

#stored as a variable
x_test = lambda x,y,z: x * y * z

print(x_test(2,4,10))
```

#### Passing a Lambda into a Function

```python
def multiply(f, num):
    """
        f expects a lambda function
        nm expects a number
    """
    
    return f(num)

multiply(lambda x: x*x, 5)
```

#### Returning a Lambda from a Function

```python
# regular defined function
def multiply_four(num):
    return num * 4

# function within a function
def return_func():
    test = 1
    def multiply (num):
        return num * 2
    return multiply

f_return = return_func()

# print(return_func())
print(f_return(4))

# lambda function returned from regular function
def return_lam(a, b):
    return lambda y, z: a * b - y * z

r_lamb = return_lam(6, 7)

print(r_lamb(4, 5))

# note: lambdas can be used as helper functions within other functions
# so that defining another function outside of a function isn't necessary 
# (keeps thing condensed)

print(return_lam(6,6)(5,5))
```

#### If Statements within Lambdas

```python
# lambda x : True if (condition) else False

greater_ten = lambda num : num * 2 if num > 10 else num + 2

print(greater_ten(8))
print(greater_ten(12))
print(greater_ten(10))
```

## Map <br>
<p>The map function allows you to iterate over an entire list while running a function on each item of the list. This is why the map function works well with lambda's, because it simplifies things and you write less lines of code.<br>The syntax for a map function is "map(function to be used, list to be used)"<br>However, you must be careful, as the map function returns a map object, not a list. To turn it into a list we use the list() type conversion.</p>

#### Syntax

```python
# map (function, iterable(list, dict, tuple, set))
# normally the map works with a pre-defined fuction - but you can use lambdas too

# func to use in map()
#(this example uses a funct taking 2 params, sighlty more compliated first example)

def squared(num1, num2):
    if num1 < 10 and num2 < 10:
        return num1 ** 2, num2 ** 2
    else:
        return num1, num2
    
nums = [4,11,25,3,15,20]
nums2 = [4,10,3,5,6]

squared_nums_map = map(squared, nums, nums2)
print(list(squared_nums_map))

```

#### Using Lambda's with Map

```python
# map(lambda x: x + 2, list)

squared_nums_lamb = list(map(lambda x,y: (x**2, y**2) if x < 10 and y < 10 else (x,y), nums, nums2))
print(squared_nums_lamb)
```

## Filter() <br>
<p>Filter's are similar to the map function, where you're able to pass a function argument and a list argument and filter out something from the list based on the conditions passed. Similar to the map function, it returns a filter object, so you need to type convert it to a list()</p>

#### Syntax

```python
names = ['Bob', 'Andy', 'Buzz', 'Angelica', 'Lil']

def a_names(name):
    if name[0].lower() == 'a':
        return True
    else:
        return False

new_names_list = list(filter(a_names, names))
print(new_names_list)
    
```

#### Using Lambda's with Filter()

```python
new_names_lamb = list(filter(lambda name: True if name[0].lower() == 'a' else False, names))
print(new_names_lamb)
```

## Recursion <br>
<p>Recursion means that a function is calling itself, so it contanstly executes until a base case is reached. It will then push the returning values back up the chain until the function is complete. A prime example of recursion is computing factorials... such that 5! (factorial) is 5*4*3*2*1 which equals 120.</p>

#### Implementing a Base Case

```python
#the below solved alternatively without using recursion 
def add_nums(num):
    return reduce(lambda x,y : x+y, [i for i in range(num+1)])
add_nums(5)
```

```python
#using recursion to solve

def addNums(num):
    # set base case
    if num <= 1:
        print("addNums(1) = 1")
        return num 
    
    # recursive call
    else:
        print(f"addNums({num}) = {num} + addNums({num - 1})")
        return num + addNums(num - 1)
    
addNums(5)
```

#### Writing a Factorial Function

```python
# 5! = 5*4*3*2*1

def factorial(num):
    #base case
    if num <= 1:
        return 1
    
    #recursive call
    else:
        return num * factorial(num - 1)
    
print(factorial(5))
```

## Generators <br>
<p>Generators are a type of iterable, like lists or tuples. They do not allow indexing, but they can still be iterated through with for loops. They are created using functions and the yield statement.</p>

#### Yield Keyword <br>
<p>The yield keyword denotes a generator, it doesn't return so it won't leave the function and reset all variables in the function scope, instead it yields the number back to the caller.</p>

```python
#the long way via function + for loop
def square_nums(nums):
    result = []
    for num in nums:
        result.append(num**2)
    return result

squares = square_nums([1,2,3,4,5])
print(squares)
```

```python
#the generator way, using 'yield'
def square_nums(nums):

    for num in nums:
        yield num**2

squares = square_nums([1,2,3,4,5])
print(squares)

for square in squares:
    print(square)
```

```python
#the generator way
def square_nums(nums):

    for num in nums:
        yield num**2

squares = square_nums([1,2,3,4,5])
print(squares)

# print(list(squares)) - this line will conert squares into a usable list

print(next(squares))  #the 'next' function can parce through generator values 
print(next(squares))
print(next(squares))
print(next(squares))
print(next(squares))
```

### Exercise #1 <br>
<p>Filter out all of the empty strings from the list below</p>

`Output: ['Argentina', 'San Diego', 'Boston', 'New York']`

```python
places = [" ","Argentina", " ", "San Diego","","  ","","Boston","New York"]

#-----------------------for loop solution---------------------

def real_places(place):
    if len(place.strip()) > 0:
        return True
    else:
        return False

print("Using a 'for loop':")
print(list(filter(real_places, places)))

#------------------------lambda solution------------------------

print("\nUsing 'lambda':")
print(list(filter(lambda place: True if len(place.strip()) > 0 else False, places)))
```

### Exercise #4 <br>
<p>Write a recursion function to perform the fibonacci sequence up to the number passed in.</p>

`Output for fib(5) => 
Iteration 0: 1
Iteration 1: 1
Iteration 2: 2
Iteration 3: 3
Iteration 4: 5
Iteration 5: 8`

```python
def fib_calc(n):
    # Set base cases for recusive function
    if n == 0:
        return 0
    elif n == 1:
        return 1
    #recursive code
    else:
        return fib_calc(n-1) + fib_calc(n-2)

def fib(n):
    for i in range(0, n+1):
        print(f"Iteration {i}: {fib_calc(i)}")

print(fib(12))
```
